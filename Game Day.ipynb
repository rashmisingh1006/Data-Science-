```python
import pandas as pd
data_frame = pd.read_excel('data.xlsx')
data_frame.shape
```

```python
df1 = pd.DataFrame(data_frame, columns=['Time', 'Quarter', 'Ad', 'Ad_List', 'Brand', 'Keywords'])
df1
```

```python
df1['Quarter'] = df1['Quarter'].map(str)
```

```python
df1["Quarter"].hist()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt 
New_colors = ['green','blue','purple','brown']
fig = plt.gcf()
fig.set_size_inches(18.5,10.5)
plt.bar(df1['Brand'], df1['Quarter'],color = New_colors)
plt.title('Brand vs Quarter - Super Bowl Game 2022', fontsize=14)
plt.xlabel('Brand', fontsize=14)
plt.ylabel('Quarter',fontsize=14)
plt.show()
plt.xticks(rotation=90)
plt.xticks(range(76))
plt.subplots_adjust(bottom=0.1)
```

```python
import matplotlib.pyplot as plt 
data = plt.scatter(df1['Keywords'],df1['Brand'], c ='b', s =20)
plt.xticks(rotation=90)
plt.xticks(range(76))
plt.xticks(fontsize=5)
plt.yticks(fontsize=4)
plt.subplots_adjust(bottom=0.1)
```
